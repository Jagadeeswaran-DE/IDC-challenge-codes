# Explain query

```python
%sql
explain
select * from ecom_idc.silver.events where event_type = 'purchases';
```

# Partitioned table

```python
%sql
create table if not exists ecom_idc.silver.events_part
using delta
partitioned by (event_date,event_type)
as select * from ecom_idc.silver.events;
```

```python
%sql
show partitions ecom_idc.silver.events_part
```

# Optimize

```python
%sql
OPTIMIZE ecom_idc.silver.events_part ZORDER BY (user_id,product_id);
```
