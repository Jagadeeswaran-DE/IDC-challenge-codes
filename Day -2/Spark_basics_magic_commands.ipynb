```python
%sql
create or replace temp view ecom_oct
using csv 
options(
path '/Volumes/idc/idc_kaggle/ecom_data/2019-Oct.csv'
,header 'true',
inferSchema 'true'
);
```

```python
%sql
SELECT event_type, brand, price
FROM ecom_oct;
```

```python
%sql
select count(*) from 
ecom_oct where price>100;
```

```python
%sql
select brand,count(*) from 
ecom_oct group by brand;
```

```python
%sql
select brand ,count(brand) from ecom_oct group by brand order by count(brand) desc limit 5;
```
