```python
%sql
WITH daily AS (
  SELECT 
    event_date, 
    SUM(revenue) AS rev 
  FROM 
    ecom_idc.gold.prod
  GROUP BY 
    event_date
)
SELECT 
  event_date, 
  rev, 
  AVG(rev) OVER (
    ORDER BY event_date 
    ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
  ) AS ma7
FROM 
  daily;
```

```python
%sql
SELECT 
  category_code, 
  SUM(views) AS total_views,
  SUM(purchases) AS total_purchases,
  ROUND(SUM(purchases) * 100.0 / SUM(views), 2) AS conversion_rate
FROM 
  ecom_idc.gold.prod
GROUP BY 
  category_code
```

```python
%sql
SELECT
  CASE WHEN cnt >= 10 THEN 'VIP'
       WHEN cnt >= 5 THEN 'Loyal'
       ELSE 'Regular' END as tier,
  COUNT(*) as customers,
  AVG(total_spent) as avg_ltv
FROM (SELECT user_id, COUNT(*) cnt, SUM(price) total_spent
      FROM ecom_idc.silver.events WHERE event_type='purchase' GROUP BY user_id)
GROUP BY tier;
```
